# From econonomist.com take out all the link with article, (not advertasing, inscription...)



```python
## import libraries
from google.colab import drive
drive.mount("/content/drive")
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from datetime import date
import os

today = date.today()

## root directory


## create a directory
def dir_create(d): # d- int position on list_web
    dir_root = f"/content/drive/MyDrive/Colab Notebooks/MIs proyectos/Web Scraper News/"
    try: os.mkdir(dir_root+ list_web[d]["site"]+f"/{today}")
    except FileExistsError:
        pass
    return dir_root + list_web[d]["site"] + f"/{today}"

   
def economist_art():
    print("Storage the economist in your computer")
    list_web_pos = 0 # position of The economist in list_web


    # try: 
    dir_create(0)

    # type of resolution
    img_resolution = 2 #  0,2,4,6,8,10,12,14 or 16 

    ## to export the date
    today = date.today()

    ## get the entry web
    url1 = "https://www.economist.com/"
    html_content = requests.get(url1).text
    soup = BeautifulSoup(html_content)

    ## take out  only the links with news from the web economist.com
    link_soap = re.findall('"(.*?)"', soup.text)  # get all the strings between "qoutation mark"
    link_soap = [i for i in link_soap if i[:26] == "https://www.economist.com/"]  # get only the string that contain www.economist.com
    link_ = [i for i  in link_soap if (i[-4] != ".") and (i[-1] in [chr(i) for i in range(97,123)]) and (i[-4:] != "-api")] #clean the list

    ## clean a little more link_url_news
    link_url_news = []
    for i in link_:
        for z in i:
            if str.isnumeric(z) and i not in link_url_news:
                # if i not in link_url_news:
                link_url_news.append(i)

    # create a list of dictionary that contain the url, title, subtitle and body
    list_content = []
    for link in link_url_news:
        dict_news = dict()
        dict_news["web"] = link
        html_content = requests.get(link).text
        soup = BeautifulSoup(html_content)
        title = soup.find_all("title")[0].text
        try: sub_title = soup.findAll("h2", class_="article__description")[0].text
        except IndexError: sub_title = None
        body_soup = soup.find_all("p", class_="article__body-text")
        body = "\n".join([i.text for i in body_soup])
        dict_news["title"] = title
        dict_news["sub_"] = sub_title
        dict_news["body"] = body
        dict_news["thema"] = re.findall("/(.*?)/",link)[1]
        dict_news["date"] = re.findall("20[0-2][0-9]+/[0-1][0-9]/[0-3][0-9]",link)[0]

        # looking for the link of the article photo
        try:photo_url = soup.find_all("img")[0]["srcset"].split()[img_resolution].split(",")[1]
        except IndexError: dict_news["photo_link"] = None
        list_content.append(dict_news)

        # Photos storage
        img_data =requests.get(photo_url).content
        with open(dir_create(0) + f'/{dict_news["title"]}.jpg', 'wb') as handler:
            handler.write(img_data)

    ## Dataframe storage
    df = pd.DataFrame(list_content)
    df.to_csv(dir_create(list_web_pos)+f"/ {today}.csv", index=False)
    print("All the info is in your computer now")

##  




## list of web with newspapers to download

list_web = [{"url" : "https://www.economist.com/",               "site" : "The Economist"}, 
            {"url" : "https://www.economist.com/weeklyedition/", "site" : "The Economist Weekly edition"}, 
            {"url" : "https://elpais.com/",                      "site" : "El País"},
            ]

## return the url root to downlad

def return_url():
    range_number = list(range(len(list_web)))
    input_new_print = "Choose:\n"
    for i in range_number: input_new_print += f'{i} for for {list_web[i]["site"]} \n'
    y = int(input(input_new_print + "\n---------->"))
     # numbers of webs of newspapers

    if y not in range_number:
        print(f"Please, enter an integer number between 1 and {len(list_web)}")
        return return_url()
    else:
        print(f'You has choose {list_web[y]["site"]}, maybe take a couple of minutes, depend of your internet conection')
        return list_web[y]["url"]

if return_url() == list_web[0]["url"]:
    economist_art()
else: print("Coming soon... sorry")
```
